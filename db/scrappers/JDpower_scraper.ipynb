```python
import re
import requests
from bs4 import Tag, NavigableString, BeautifulSoup
import pandas as pd
#import csv

URL = "https://www.jdpower.com/cars/consumer-reviews"
re = requests.get(URL)
soup = BeautifulSoup(re.content, 'html.parser')
```

```python
years = []
for year in soup.find(title = "Year"):
    if isinstance(year, Tag):
        if year.text.isdigit():
            years.append(year.text)
            
print(years)
```

```python
makes = []
for make in soup.find_all(title = "Make")[1]:
    if isinstance(make, Tag):
#         print(make)
        makes.append(make.text)

makes = makes[1:]
print(makes)
```

```python
ratings = pd.DataFrame(columns=['Year', 'Make', 'Model', 'Overall', 'Quality_And_Reliability', 'Driving_Experience', 'Resale'])
for year in years:
    #print(year)
    for make in makes:
        #print(make)
        temp_URL = "https://www.jdpower.com/cars/" + year +"/" + make + "/"
        temp_re = requests.get(temp_URL)
        temp_soup = BeautifulSoup(temp_re.content, 'html.parser')
        
        #models = []
        for trim in temp_soup.find_all(class_='trim-name'):
            try:
                model = trim.text.strip()
                #models.append(trim)
                final_URL = "https://www.jdpower.com"+trim['href']
                final_re = requests.get(final_URL)
                final_soup = BeautifulSoup(final_re.content, 'html.parser')
                print(final_URL)
            except:
                print("No link!")
                continue
                
            values = []

            try:
                for i in final_soup.find_all(class_='content')[:4]:
                    values.append(i.find(class_='value').text.strip())
                print([year,make,model,values[0],values[1],values[2],values[3]])
            except:
                print("Data is missing!")
                continue
                
            ratings_temp = pd.DataFrame(data=[[year,make,model,values[0],values[1],values[2],values[3]]],columns=['Year', 'Make', 'Model', 'Overall', 'Quality_And_Reliability', 'Driving_Experience', 'Resale'])
            ratings = pd.concat([ratings,ratings_temp])
            
        
```

```python
ratings.to_csv("ratings.csv", encoding='utf-8')
```
