```python
import pandas as pd
```

```python
standard = pd.read_csv('us_car_models.csv')
standard
```

```python
ratings = pd.read_csv('jdpower_ratings_all.csv',usecols=[1,2,3,4,5,6,7])
```

```python
ratings
```

```python
og = []
for i in range(len(standard)):
    og.append((standard.loc[i].make.upper(),standard.loc[i].model.upper()))
og = list(set(og))
```

```python
jd = []
for i in range(len(ratings)):
    jd.append((ratings.loc[i].Make.upper(),ratings.loc[i].Model.upper()))
jd = list(set(jd))
```

```python
missing = []
for i in jd:
    if i not in og:
        missing.append(i)
```

```python
models = []
for i in missing:
    models.append(i[0])
models = list(set(models))
models
```

```python
for model in models:
    print(model)
    print('missing make, model')
    for miss in missing:
        if miss[0] == model:
            print(miss)
    print('possible make, model')
    for i in og:
        if i[0] == model:
            print(i)
```

```python
for i in range(len(ratings)):
    if ratings.loc[i].Make.upper() == 'DODGE' and ratings.loc[i].Model.upper() == 'GRAND CARAVAN CARGO VAN':
        ratings.set_value(i,'Model','GRAND CARAVAN CARGO')
    elif ratings.loc[i].Make.upper() == 'DODGE' and ratings.loc[i].Model.upper() == 'GRAND CARAVAN':
        ratings.set_value(i,'Model','GRAND CARAVAN PASSENGER')
    elif ratings.loc[i].Make.upper() == 'CHRYSLER' and (ratings.loc[i].Model.upper() == 'TOWN AND COUNTRY LWB' 
                                                        or ratings.loc[i].Model.upper() == 'TOWN AND COUNTRY SWB'):
        ratings.set_value(i,'Model','TOWN AND COUNTRY')
    elif ratings.loc[i].Make.upper() == 'CHRYSLER' and ratings.loc[i].Model.upper() == 'PT CRUISER CLASSIC':
        ratings.set_value(i,'Model','PT CRUISER')
    elif ratings.loc[i].Make.upper() == 'CHRYSLER' and ratings.loc[i].Model.upper() == 'TOWN AND COUNTRY':
        ratings.set_value(i,'Model','TOWN & COUNTRY')
    elif ratings.loc[i].Make.upper() == 'SUBARU' and (ratings.loc[i].Model.upper() == 'LEGACY SEDAN'  
                                                      or ratings.loc[i].Model.upper() == 'LEGACY WAGON'): 
        ratings.set_value(i,'Model','LEGACY')
    elif ratings.loc[i].Make.upper() == 'SUBARU' and (ratings.loc[i].Model.upper() == 'IMPREZA SEDAN WRX'  
                                                      or ratings.loc[i].Model.upper() == 'IMPREZA SEDAN'  
                                                      or ratings.loc[i].Model.upper() == 'IMPREZA WAGON' 
                                                      or ratings.loc[i].Model.upper() == 'IMPREZA WAGON WRX'):
        ratings.set_value(i,'Model','IMPREZA')
    elif ratings.loc[i].Make.upper() == 'SUBARU' and ratings.loc[i].Model.upper() == 'WRX STI': 
        ratings.set_value(i,'Model','WRX')
    elif ratings.loc[i].Make.upper() == 'SUBARU' and ratings.loc[i].Model.upper() == 'XV CROSSTREK HYBRID': 
        ratings.set_value(i,'Model','XV CROSSTREK')
```
